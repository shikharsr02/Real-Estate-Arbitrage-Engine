```python
df = pd.read_csv(r'C:\Users\sahil\OneDrive\Desktop\Own Projects\Residential\real_estate_Cleaned.csv')
```

```python
#dropping not needed columns
df.drop(columns=['Unnamed: 0', 'carpet.area'],inplace = True)
```

```python
#filling the missing data
#filling unimportant data with zero like parking, balconies,etc
#removing data with important data missing like price

df['parking']=df['parking'].fillna(0)
df['balcony']=df['balcony'].fillna(0)

df.dropna(subset=['Price', 'total_area'], inplace=True)
```

```python
#Converting parking data into integer from covered/open to integer only

#Extracting integer part of the string

df['parking'] = df['parking'].str.extract(r'(\d+)')

#Filling all NA with 0

df['parking']=df['parking'].fillna(0)

#Converting string to integer using astype

df['parking']=df['parking'].astype(int)
```

```python
#categorizing property type using listing name

df['Property_Type'] = np.where(df['Name'].str.contains('Villa'), 'Villa',
                      np.where(df['Name'].str.contains('House'), 'House', 'Flat'))
```

```python
#Rounding off total area to 2 decimal places
df['total_area']= np.round(df['total_area'],2)
```

```python
#standardization total area in sq feet

df['carpet_area(in sqft)'] = np.where(df['Area_Unit']=='sqyrd',df['carpet_area_sqft']*9,
                                 np.where (df['Area_Unit']=='sqm', df['carpet_area_sqft']*10.764,
                                  df['carpet_area_sqft']))

```

```python
# Calculating Z-score for price per sqft within each city
df['Z_Score'] = df.groupby('city')['Rate_per_sqft'].transform(lambda x: (x - x.mean()) / x.std())

# Keep only the rows that fall within 3 standard deviations (removes extreme outliers)
df_clean = df[df['Z_Score'].between(-3, 3)]
```

```python
df_clean.to_csv('Real_Estate_Cleaned_3.csv',index=False)
```
